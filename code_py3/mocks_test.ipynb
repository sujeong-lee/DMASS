```python
import numpy 
import pylab as plt 
import scipy.optimize
import scipy

file_temp = "/fs/scratch/PCON0008/warner785/bwarner/mocks/mocks_set20_pca{0}.txt"
outdir = '/fs/scratch/PCON0008/warner785/bwarner/'
cov_template = 'cov_fixed3_{0}'

delta = 1e-10

#loop over 50 PCA maps
for ipca in range(50):
    data = numpy.loadtxt(file_temp.format(ipca))
    cov = numpy.cov(data,rowvar=False)

    #try:
      #  scipy.linalg.cholesky(cov)
      #  print(f'ipca={ipca}. No need for delta')
    #except scipy.linalg.LinAlgError:

        #add a small value to the diagonal
    numpy.fill_diagonal( cov, cov.diagonal()+delta )
    try:
        scipy.linalg.cholesky(cov)
        print(f'ipca={ipca}. delta={delta} fixed the matrix')
    except scipy.linalg.LinAlgError:
        print(f'ipca={ipca}. delta={delta} did not fix the matrix!!!!!!!!!!!')
    cov_keyword = cov_template.format(ipca)
    np.savetxt(outdir+cov_keyword+'.txt', cov)
```

```python
#What you have is the number density array (12 data points) for each mock and PCA map
#When you put all the number density arrays for a given PCA map into np.cov it will output 
#the covariance matrix for that map. you dont need to add anything to this

#A bit about what the covariance matrix means: Your matrix should be a 12x12 matrix. 
#The diagonal of the covariance matrix is the variance of the number density data points 
#(not the variance on the mean of the mocks).  And the sqrt of the diagonal is the standard deviation. 
#The off-diagonal terms describe how correlated each data point is with another 
#(i.e. for a positively correlated data, if the true value of a given data point is high, 
#it is more likely that another the other data points are high)

#You should have 50 covariance matrices in total, one for each PCA map
```
